### Diabetes Insipidus
Diabetes Insipidus: Key Points Summary

- Causes & forms: 
  - Central DI: Lack of ADH production from the hypothalamus.
	- Nephrogenic DI: Kidneys don’t respond to ADH.
  - (also Gestational DI: ADH is degraded by placental enzymes & Dipsogenic DI: Dysfunction of thirst mechanism. -> less common)
- Disease Progression: Progressive inability to concentrate urine leading to severe dehydration if untreated, can reach up to 20l 
- Early Symptoms & Biomarkers:
  - Symptoms: Excessive urination and thirst.
  - Biomarkers: Low urine osmolality and specific gravity; normal or low serum ADH.
- General Symptoms & Biomarkers:
  - Symptoms: Persistent diuresis, dehydration, hypernatremia.
  - Biomarkers: High plasma osmolality, low urine osmolality, possibly elevated serum sodium.

#### Traditional Diagnosis of DI
Testing for DI is expensive & difficult, as it is not easy dto differentiate between central DI and primary polydipsia
-> difference in pathophysiology: in the disease progression of primary polydipsia, the circulus vitiosus starts with development of excessive thirst, which then leads to excessive urination to excrete the fluids the person drank, while in DI, the disease begins with polyuria as the kidneys cannot concentrate the urine anymore and the polydipsia is secondary to the fluid loss

Diagnostics:
- Medical History / Examination
- Urine: Osmolality, Urine Specific Gravity
- Blood: Serum Osmolality, Blood Sodium Levels (secondary to the Polyuria -> concentration due to dehydration)
- Water Deprivation Test: really straining and uncomfortable for the patients!
- **Differentiation between central and nephrogenic DI**: Vasopressin (ADH) Test
  - in central DI, the administration will lead to reduced polyuria, as the pathophysiology is that not enough ADH is produced and secreted from the hypophysis, which secondarily leads to increased urine production because there is not enough ADH that increases fluid re-absorption
  - while in nephrogenic DI there will be little to no change with ADH administration, because the pathophysiology is that the receptors for the ADH in the kidneys are not reacting to the ADH
- hypertonic saline infusion test:
- imaging: MRI of brain -> hypothalamus or pituitary gland

#### Hypertonic Saline Test
How the Hypertonic Saline Test Works:

- Administration of Hypertonic Saline: The test involves intravenously administering hypertonic saline (a highly concentrated salt solution) to increase plasma osmolality. This rise in plasma osmolality is designed to stimulate the pituitary gland to release ADH.
- Monitoring: Throughout the test, the patient’s blood sodium levels and osmolality are closely monitored. Additionally, the levels of ADH in the blood are measured at intervals.
- Response Evaluation: The normal response to hypertonic saline infusion is an increase in the secretion of ADH as the body reacts to the induced hyperosmolality. By measuring ADH levels and how they correlate with changes in blood osmolality, healthcare providers can assess the responsiveness and functionality of the pituitary in regulating ADH.

Uses of the Test:

- Differentiating Between Types of Diabetes Insipidus: The test can help differentiate central diabetes insipidus (where ADH production is deficient) from nephrogenic diabetes insipidus (where the kidneys do not respond to ADH). In central DI, ADH levels will not rise appropriately with increased plasma osmolality, whereas in nephrogenic DI, ADH may be released but the kidneys fail to respond to it.
- Diagnosing SIADH: In cases of SIADH, the test helps confirm excessive ADH secretion even when plasma osmolality is low or normal, indicating inappropriate ADH release.

careful: risks associated with inducing hypernatremia (high blood sodium levels) and its potential complications.

### Identifying & Removing outliers

```python
def analyze_outliers(df):
    outliers_dict = {}
    
    for column in df.columns:
        # Create boxplot
        plt.figure(figsize=(8, 6))
        plt.boxplot(df[column].dropna(), vert=False)
        plt.title(f'Boxplot of {column}')
        plt.xlabel(column)
        plt.show()  # Display the plot
        
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        
        # Calculate the IQR
        IQR = Q3 - Q1
        
        # Determine outlier criteria
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Identify outliers
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        
        # Save outliers in the dictionary
        outliers_dict[column] = outliers

    return outliers_dict

# Call the function
outliers_dict = analyze_outliers(df)

# Display the outliers for each column
for column, outliers in outliers_dict.items():
    print(f'Outliers for {column}:')
    print(outliers)
```

## Imputation

Trying different methods of imputation (KNN & RandomForest Imputation with Iterative Imputer) -> see which provides the better result (= the lower mean squared error)

```python
def evaluate_imputation(X, y, imputer, model):
    pipeline = Pipeline([
        ('imputer', imputer),
        ('scaler', StandardScaler()),  # Scaling is applied after imputation
        ('regressor', model)
    ])
    
    # Set up cross-validation
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    
    # Cross-validate the pipeline
    scores = cross_val_score(pipeline, X, y, scoring='neg_mean_squared_error', cv=kf)
    
    # Return average MSE
    return np.mean(-scores)
```

```python
# Initialize the KNN imputer
knn_imputer = KNNImputer(n_neighbors=5)

# Initialize the Iterative Imputer using RandomForest
rf_imputer = IterativeImputer(estimator=RandomForestRegressor(n_estimators=10, random_state=42), random_state=42)

# Initialize the regression model
model = RandomForestRegressor(n_estimators=100, random_state=42)
```

### Evaluating Imputation for DF with engineered feature

```python
# Evaluate KNN imputer
mse_knn = evaluate_imputation(X_train_eng, y_train_eng, knn_imputer, model)
print(f"KNN Imputer MSE: {mse_knn}")

# Evaluate Random Forest imputer (IterativeImputer)
mse_rf = evaluate_imputation(X_train_eng, y_train_eng, rf_imputer, model)
print(f"Random Forest Imputer MSE: {mse_rf}")

# Determine the best imputer
best_imputer = 'KNN' if mse_knn < mse_rf else 'Random Forest'
print(f"Best imputer selected based on cross-validated MSE: {best_imputer}")
```

```python
import shap

# Assuming 'model' is your trained classifier and X_val_imputed_single is your feature DataFrame
explainer = shap.Explainer(best_pipeline.predict, X_val_imputed_single)

# Calculate SHAP values for false positives and false negatives
shap_values_fp = explainer(false_positives.drop(['Actual', 'Predicted'], axis=1))
shap_values_fn = explainer(false_negatives.drop(['Actual', 'Predicted'], axis=1))
```

```python
# Summary plot for false positives
shap.summary_plot(shap_values_fp, false_positives.drop(['Actual', 'Predicted'], axis=1))

```

```python
# Summary plot for false negatives
shap.summary_plot(shap_values_fn, false_negatives.drop(['Actual', 'Predicted'], axis=1))
```

```python
# Generate a waterfall plot for each false positive
for i in range(len(false_positives)):
    print(f"Waterfall Plot for False Positive Index: {false_positives.index[i]}")
    shap.plots.waterfall(shap_values_fp[i], max_display=10)
    plt.show()  # Make sure to display the plot if not automatically shown

# Generate a waterfall plot for each false negative
for i in range(len(false_negatives)):
    print(f"Waterfall Plot for False Negative Index: {false_negatives.index[i]}")
    shap.plots.waterfall(shap_values_fn[i], max_display=10)
    plt.show()  # Make sure to display the plot if not automatically shown
```
